```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
import cv2
import os
import numpy as np
import tensorflow as tf
from sklearn.utils import shuffle
from keras.layers import Flatten
import tensorflow_probability as tfp
from tensorflow.keras.layers import Flatten, Dense
from keras.layers import Dropout
from keras.layers import SimpleRNN, Dense, Flatten
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GRU, Dense, Dropout, Flatten
from tensorflow.keras.optimizers import Adam




X_train = []
Y_train = []
image_size = 150
labels = ['glioma_tumor', 'meningioma_tumor', 'no_tumor', 'pituitary_tumor']

# Load and preprocess your data
for i in labels:
    folderPath = os.path.join('archive/train', i)
    for j in os.listdir(folderPath):
        img = cv2.imread(os.path.join(folderPath, j))
        img = cv2.resize(img, (image_size, image_size))
        X_train.append(img)
        Y_train.append(i)

for i in labels:
    folderPath = os.path.join('archive/test', i)
    for j in os.listdir(folderPath):
        img = cv2.imread(os.path.join(folderPath, j))
        img = cv2.resize(img, (image_size, image_size))
        X_train.append(img)
        Y_train.append(i)

X_train = np.array(X_train)
Y_train = np.array(Y_train)

# Shuffle the data
X_train, Y_train = shuffle(X_train, Y_train, random_state=101)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_train, Y_train, test_size=0.1, random_state=101)

# Preprocess the labels
label_to_index = {label: index for index, label in enumerate(labels)}
y_train = np.array([label_to_index[label] for label in y_train])
y_train = tf.keras.utils.to_categorical(y_train, num_classes=len(labels))

y_test = np.array([label_to_index[label] for label in y_test])
y_test = tf.keras.utils.to_categorical(y_test, num_classes=len(labels))




model_rnn = Sequential([
    GRU(64, input_shape=(X_train_rnn.shape[1], X_train_rnn.shape[2]), activation='tanh', recurrent_activation='sigmoid'),
    Dropout(0.5),
    Dense(64, activation='relu'),
    Dense(4, activation='softmax')  # Assuming 4 output classes for the tumor types
])

# Compile the model
model_rnn.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the RNN model
history_rnn = model_rnn.fit(X_train_rnn, y_train, epochs=20, validation_split=0.1)



```
